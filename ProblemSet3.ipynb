### Question 1: Occupations

##### Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

##### import the dataset & Assign it to a variable called users

```python
users = pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', '|')
users
```

##### Discover what is the mean age per occupation

```python
users[["occupation","age"]].groupby("occupation").mean()
```

##### Discover the Male ratio per occupation and sort it from the most to the least

```python
tmp_df = users[["occupation","gender","user_id"]].groupby(["occupation", "gender"]).count()
tmp_df
```

```python
p_table = pd.pivot_table(tmp_df,'user_id','occupation','gender').reindex()
p_table['M_ratio'] = p_table['M']/p_table['F']
p_table.sort_values(by=['M_ratio'], ascending=False)
```

##### For each occupation, calculate the minimum and maximum ages

```python
users[["occupation","age"]].groupby("occupation").agg(['min', 'max'])
```

##### For each combination of occupation and sex, calculate the mean age

```python
tmp_df = users[["occupation","gender","age"]].groupby(["occupation", "gender"]).mean()
tmp_df
```

##### For each occupation present the percentage of women and men

```python
tmp_df = users[["occupation","gender","user_id"]].groupby(["occupation", "gender"]).count()
tmp_df
```

```python
p_table = pd.pivot_table(tmp_df,'user_id','occupation','gender').reindex().fillna(0)
p_table['F_percentage'] = (p_table['F']/(p_table['F']+p_table['M']))*100
p_table['M_percentage'] = (p_table['M']/(p_table['F']+p_table['M']))*100
p_table
```
